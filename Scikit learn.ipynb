En este notebook vamos a ver dos ejemplos, un primer ejemplo de regresión, que consiste en la estimación de un valor númerico y un segundo caso de clasificación, que consiste en asignarle una clase o etiqueta a cada elemento.

## Regreción
En este ejemplo vamos a realizar en ejemplo de regresión, en el cual a apartir de unas caracteristicas (o *features*) vamos a estimar un determiando valor, predicción.

Para este ejemplo usaremos el conjunto de datos de precios de la vivienda publicado en [Kaggle](https://www.kaggle.com/harlfoxem/housesalesprediction).

Comencemos cargando los datos.

```python
import pandas as pd

preciosCasas = pd.read_csv('./recursos/kc_house_data.csv') 
```

Mostramos los registros cargados

```python
preciosCasas.head(10)
```

```python
preciosCasas.columns
```

### Selección de caracteristicas
Una vez cargados los datos seleccionaremso aquellas variables que usaremos para la predicción. Esta tarea es una labor importante para el correcto funcioanmiento de los modelos, y la cual no se encuentra en el ámbito de este curso, por lo que seleccionaremos las variables principales sin la utilización de técnicas de seleccion de variables, lo cual es un campo de trabajo por si solo.

Para este ejemplo usaremos las variables: bedrooms, bathrooms, sqft_living, sqft_lot, floors, lat y long. Este conjunto se suele denominar como X.

```python
X = preciosCasas[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'lat', 'long']]
X.head(10)
```

```python
X.describe()
```

### Variable a estimar
A continuación obtenemos la variable la cual queremos estimar, es este caso el valor de price. Este valor se suele denominar Y

```python
Y = preciosCasas.price
Y.describe()
```

#### Datos de entrenamiento y test
Antes del entrenamiento de los modelos, es necesario partir los datos almenos en entrenamiento y test, ya que en el caso de entrenar y probar los modelos con el mismo conjunto no se obtendrá un error de generalización correcto, ya que el modelo "ya conoce" esos datos y se le estimará un funcionamiento mejor que el que tendrá cuando entre en funcionamiento

A continuación partimos los datos, dejando un 80% de los datos para entrenamiento un 20% de test.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.2)
```

### Entrenamiento del modelo
A continuación vamos a realizar el entrenamietno del modelo a. En este caso hemos seleccionado [SVR](https://scikit-learn.org/stable/modules/svm.html#regression) siguiendo la [guía de seleccion de modelos](https://scikit-learn.org/stable/modules/svm.html#regression).

Para esta tarea vamos a usar los datos de entrenamiento, **nunca usaremso los de test**.

Vamos a ello, a continuación vamos a entrenar nuestro modelo con los valores de X de test e Y de test seleccionado anteriormente.

```python
from sklearn import svm

modelo = svm.SVR()
modelo.fit(x_train, y_train)
```

A continaución vamos a obtener la precicción para los diez primeros elemento del conjunto de test

```python
print(x_test.head(10))
print("Predicciones")
print(modelo.predict(x_test.head(10)))
```

### Validación del error
Una vez tenemos entrenado nuestro modelo, vamos a validar cuanto de bueno es. Para ello realizaremos la predicción con los datos de test y comprobaremos su salida con la real almacenada en la varible y_test. Esta comprobación se puede realizar de múltiples maneras, en este ejemplo vamos a usar el error cuadrático media y el R2.


```python
from sklearn import metrics

y_prediccion = modelo.predict(x_test)

mse = metrics.mean_squared_error(y_test, y_prediccion)
r2 = metrics.r2_score(y_test, y_prediccion)

print("Metricas de error")
print("------------------")
print('MSE is {}'.format(mse))
print('R2 score is {}'.format(r2))
```

## Clasificación
Las tareas de clasificación tienen por objetivo asignar una clase o etiqueta a una determianda observacion. 

Para este ejemplo vamos a usar el cojuntto de datos [UCI](https://www.kaggle.com/ronitf/heart-disease-uci), el cual tiene por objetivo decir si una persona tiene una enfermedad o no.

Vamos a comenezar como con la regresión, cargando los datos

```python
import pandas as pd

uci = pd.read_csv('./recursos/heart.csv') 
```

Mostramos los datos cargados

```python
uci.head(10)
```

```python
uci.describe()
```

Mirando los datos y su descripción se puede ver que las variables cp, thal y slope son de tipo categorico, por lo que será necesario realizar su transformación, convirtiendo cada una de estas columnas en tantas como posibles valores tiene y poniendo ceros y unos, luego veremos el resultado.

```python
cp_t = pd.get_dummies(uci['cp'], prefix = "cp")
thal_t = pd.get_dummies(uci['thal'], prefix = "thal")
slope_t = pd.get_dummies(uci['slope'], prefix = "slope")

frames = [uci, cp_t, thal_t, slope_t]
uci = pd.concat(frames, axis = 1)
uci.head()
```

Como se puede ver en la primera fila la variable cp tiene el valor 3, por lo que las columnas obtenidas cp_1 y cp2_tiene valor 0 y la columna cp_3 tiene el valor 1.


Una vez realizamos esta trassformacion eliminamos las variables originales que hemos transformado.

```python
uci = uci.drop(columns = ['cp', 'thal', 'slope'])
uci.head()
```

Creamos las variables X e Y igual que en el caso de regresión

```python
Y = uci.target.values
X = uci.drop(['target'], axis = 1)
```

#### Datos de entrenamiento y test

Al igual que hicimso en el caso de regresion partimos el conjunto de datos en entrenemaiento y test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.2)
```

### Entrenamiento del modelo
A continuación vamos a entrenar nuestro clasificador, en este caso vamos a usar el modelo llamada [SVC](https://scikit-learn.org/stable/modules/svm.html#classification), siguiendo la [guia de selección de modelos](https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html)

```python
from sklearn import svm
clasificador = svm.SVC()
clasificador.fit(x_train, y_train)
```

```python
print("Predicciones")
print(clasificador.predict(x_test.head(10)))
print(y_test[0:10])
```

### Validación del error
Una vez tenemos entrenado nuestro modelo, vamos a validar cuanto de bueno es. Para ello realizaremos la predicción con los datos de test y comprobaremos su salida con la real almacenada en la varible y_test. Esta comprobación se puede realizar de multiples maneras, en este ejemplo vamos a usar la matriz de confusión y el accuracy.



```python
from sklearn import metrics

y_prediccion = clasificador.predict(x_test)

mc = metrics.confusion_matrix(y_test, y_prediccion)
accuracy = metrics.accuracy_score(y_test, y_prediccion)

print("Métricas de error")
print("-----------------")
print('Matriz de confusión')
print(mc)
print('Accuracy {}'.format(accuracy))
```
